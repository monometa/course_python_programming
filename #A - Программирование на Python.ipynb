### Решение задачи № 2.4.3

```python
string = 'aaaaaaaaaaaaaaadddddddddddddrrrrrrrrrrccccccccca' 
output = string[0]
count = 0
i = 0
if len(string) == 1:
    output += '1'
while i < len(string) - 1:
    if string[i] == string [i+1]:
        count += 1
    else: 
        output += str(count + 1)
        output += string[i+1]
        count = 0
    i += 1
    if i == len(string) - 1:
        output += str(count + 1)
print(output)
```

```python
#best practise
genome = 'aaaaaaaaaaaaaaadddddddddddddrrrrrrrrrrccccccccca' + ' '
count = 0
el = genome[0]
res = ''
for i in genome:       
    if el != i:
        res += el + str(count)
        count = 0
        el = i
    count += 1
print(res)
```

### Задача № 2.5.4


```python
string = [int(i) for i in input().split()]
output = []

if len(string) == 1:
    output = string
else:
    for i in range(len(string)):
        if i == len(string) - 1:
            temp = string[i-1] + string[0]
        else:
            temp = string[i-1] + string[i+1]
        output.append(temp)

print(*output)
```

```python
#len %
a = [int(i) for i in input().split()]
if(len(a) == 1):
    print(a[0])
else:
    for i in range(len(a)):
        print(a[i - 1] + a[(i + 1) % len(a)],end=' ')
```

```python
#len minus
lst= [int(i) for i in input().split()]
if len(lst) == 1:
        print(lst[0])
else:
    for i in range(len(lst)):
        print(lst[i-1] + lst[i+1-len(lst)], end=' ')
```

### Задача № 2.5.5


```python
s = [int(i) for i in input().split()]
s.sort()
res = []
for i, item in enumerate(s):
    if s.count(item) > 1 and item not in res:
        res.append(item)
#print(s)
print(*res)
```

```python
#best practise
str = [int(i) for i in input().split()]
ans = []
[ans.append(x) for x in str if x not in ans and str.count(x) > 1]
print(*ans)
```
